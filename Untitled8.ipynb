```python
from detecto import core, utils, visualize
from detecto.visualize import show_labeled_image, plot_prediction_grid
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
custom_transforms = transforms.Compose([
transforms.ToPILImage(),
transforms.Resize(900),
transforms.RandomHorizontalFlip(0.5),
transforms.ColorJitter(saturation=0.2),
transforms.ToTensor(),
utils.normalize_transform(),
])
```

```python
Train_dataset=core.Dataset("/content/gdrive/MyDrive/Newdata/train/",transform=custom_transforms)#L1
Test_dataset = core.Dataset("/content/gdrive/MyDrive/Newdata/test/")#L2
loader=core.DataLoader(Train_dataset, batch_size=2, shuffle=True)#L3
model = core.Model(["apple"])#L4
losses = model.fit(loader, Test_dataset, epochs=5, lr_step_size=5, learning_rate=0.001, verbose=True)#L5
```

```python
plt.plot(losses)
plt.show()
```

```python
model.save("model_weights.pth")
model = core.Model.load("model_weights.pth", ["apple"])
```

```python
image = utils.read_image("/content/gdrive/MyDrive/Newdata/check/images3.jpg") 
predictions = model.predict(image)
labels, boxes, scores = predictions
show_labeled_image(image, boxes, labels)
```
